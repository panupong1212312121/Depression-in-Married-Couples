From infering this Beck-Depression-Inventory-BDI , 

I decide to calculate total depression score by summation column 'B1' to 'B21' and then keep these sum value to column 'Depression Score' 

and seperate group by inferring criteria

```python
df['Depression Score'] = df.iloc[:, 6:].sum(axis=1)
df = df.drop(df.columns[6:-1], axis=1)
df = df.drop('No', axis=1)

def map_depression_score(score):
    if score <= 10:
        return 'Normal'
    elif score <= 16:
        return 'Mild mood disturbance'
    elif score <= 20:
        return 'Borderline clinical depression'
    elif score <= 30:
        return 'Moderate depression'
    elif score <= 40:
        return 'Severe depression'
    else:
        return 'Extreme depression'

# apply the function to create a new column with labels
df['Depression Label'] = df['Depression Score'].apply(map_depression_score)
df =  df.drop('Depression Score', axis=1)
df
```

# How are each feature related?

As you can see , All features in this data are qualitative so I use chi-square test 

```python
# Check assumption 

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

X = df.iloc[:,:-1]

pair_break_assump = []

def check_expected_frequency(observed_frequencies):
    # Calculate the row and column sums
    row_sums = np.sum(observed_frequencies, axis=1)
    col_sums = np.sum(observed_frequencies, axis=0)
    total = np.sum(observed_frequencies)
    
    # Calculate the expected frequencies
    expected_frequencies = np.zeros(observed_frequencies.shape)
    for i in range(observed_frequencies.shape[0]):
        for j in range(observed_frequencies.shape[1]):
            expected_frequencies[i, j] = (row_sums[i] * col_sums[j]) / total

    return expected_frequencies

# Group the data by the two variables and calculate the frequency counts

for i in range(len(X.columns)-1):
    for j in range(i+1,len(X.columns)):
        grouped = X.groupby([X.iloc[:,i], X.iloc[:,j]]).size().reset_index(name="counts")
        contingency_table = grouped.pivot(index=X.columns[i], columns=X.columns[j], values="counts").fillna(0).astype(int)

        # Extract the observed frequencies from the contingency table
        observed_frequencies = contingency_table.values

        # Check the chi-square test assumptions
        expected_frequencies = check_expected_frequency(observed_frequencies)

        name = X.columns[i] + ' & ' + X.columns[j]

        # Check if the expected frequencies are at least minimum_expected_frequency
        if np.min(expected_frequencies) < 5:
            print("Warning! : One or more expected frequencies is less than 5")
            pair_break_assump.append(name)

        print(name)

        # Create contingency_table
        fig, ax = plt.subplots(figsize=(5, 4))
        heatmap = sns.heatmap(contingency_table, annot=True, cmap="YlGnBu", fmt="d", ax=ax)
        
        # Label the rows and columns
        ax.set_xticklabels(contingency_table.columns, rotation=90)
        ax.set_yticklabels(contingency_table.index, rotation=0)
        ax.set_xlabel(X.columns[j])
        ax.set_ylabel(X.columns[i])
        ax.set_title(name)

        plt.show()
        print('-'*150)

print('pair_break_assump : \n {} \n'.format(pair_break_assump))
```

From above , no pair break assumption of chi-square test -> result after chi-square testing should be precise

```python
from scipy.stats import chi2_contingency

pair_sig = []
pair_insig = []

for i in range(len(X.columns)-1):
    for j in range(i+1,len(X.columns)):
        grouped = X.groupby([X.iloc[:,i], X.iloc[:,j]]).size().reset_index(name="counts")
        contingency_table = grouped.pivot(index=X.columns[i], columns=X.columns[j], values="counts").fillna(0).astype(int)

        # Extract the observed frequencies from the contingency table
        observed_frequencies = contingency_table.values

        chi2, p, dof, expected_frequencies = chi2_contingency(observed_frequencies)

        name = X.columns[i] + ' & ' + X.columns[j]

        if p<0.05:
            print('{:<50} has significant relation with chi-square = {} and p-value = {}'.format(name,chi2,p))
            pair_sig.append(name)
        else:
            print('{:<50} has insignificant relation with chi-square = {} and p-value = {}'.format(name,chi2,p))
            pair_insig.append(name)

        # Create expected_frequencies
        fig, ax = plt.subplots(figsize=(5, 4))
        heatmap = sns.heatmap(expected_frequencies, annot=True, cmap="coolwarm", fmt=".2f", ax=ax)
        
        # Label the rows and columns
        ax.set_xticklabels(contingency_table.columns, rotation=90)
        ax.set_yticklabels(contingency_table.index, rotation=0)
        ax.set_xlabel(X.columns[j])
        ax.set_ylabel(X.columns[i])
        ax.set_title(name)

        plt.show()
        print('-'*150)

print('pair_sig : \n {} \n'.format(pair_sig))
print('pair_insig : \n {}'.format(pair_insig))
```

From above , This data is imbalance because as you can see , There is only 9 and 2 in Column 'Severe depression' and 'Extreme depression' respectively 

so I choose StratifiedShuffleSplit to handle this problem

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

from sklearn.neural_network import MLPClassifier

from sklearn.model_selection import cross_val_score, StratifiedShuffleSplit

X = df.drop('Depression Label',axis=1)
y = df['Depression Label']

models = []
models.append(('LR', LogisticRegression(max_iter=1000)))
models.append(('kNN', KNeighborsClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVC', SVC()))
models.append(('RFC', RandomForestClassifier(random_state=0)))
models.append(('DT', DecisionTreeClassifier(random_state=0)))
models.append(('MLP', MLPClassifier(hidden_layer_sizes=16,max_iter=10000,random_state=0)))

names = []

# Using StratifiedShuffleSplit:
cv = StratifiedShuffleSplit(n_splits=10, test_size=0.2, random_state=0)
for name, model in models:
    scores = cross_val_score(model, X, y, cv=cv)
    names.append(name)
    print('{:4} -> Accuracy : {:.3f} % (Std : {:.3f} %)'.format(name, scores.mean()*100, scores.std()*100))
```

From result above , I choose SVC and MLP algorithm because that has maximum scores.mean() and minimum scores.std() among others

Try to tuning SVC model and MLPClassifier model 

```python
from sklearn.model_selection import GridSearchCV

param_dist = {
    'C': list(range(1,10)),
    'kernel': ['linear'],
    'shrinking': [True, False],
    'gamma': ['auto', 1, 0.1],
    'coef0': [0.0, 0.1, 0.5]
}

# Define the SVC model
SVC = SVC()

# Run GridSearchCV to find the best hyperparameters
grid_search = GridSearchCV(
    SVC, 
    param_dist, 
    cv=StratifiedShuffleSplit(n_splits=10, test_size=0.2, random_state=0),
    verbose=1)

grid_search.fit(X, y)

print(f"Best parameters: {grid_search.best_params_}")
print(f"Best score: {grid_search.best_score_:.3f}")
```

Check for overfitting in these two models, 

I plot the learning curves to see,

If there is a large gap between the training and test scores (The training score continues to increase while the test score levels off) 

It could indicate overfitting.

```python
from sklearn.model_selection import learning_curve

from sklearn.svm import SVC

Model_name  = [SVC(C = 1, coef0 = 0.0, gamma ='auto', kernel= 'linear',shrinking= True) ,
                MLPClassifier(alpha=0.0001,hidden_layer_sizes=8,max_iter=10000,random_state=0)]

Model_title = ['SVC','MLP']

for i in range(2):
    cv=StratifiedShuffleSplit(n_splits=10, test_size=0.2, random_state=0)
    
    train_sizes, train_scores, test_scores = learning_curve(Model_name[i],X, y, cv=cv, train_sizes=np.linspace(.1, 1.0, 5))

    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std, train_scores_mean + train_scores_std, alpha=0.1, color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std, test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r", label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g", label="Cross-validation score")
    plt.title(Model_title[i])
    plt.ylabel('Accuracy')
    plt.xlabel('Sizes')
    plt.legend(loc="best")
    plt.show()
```

From above , as you can see there is no large gap between these two lines from 2 models so I summarize that my models are not overfitting

MLPClassifier :

I choose run loop 1000 times (random_state = 0,1,2,....,999) 

And then find avg_importance and avg_coefficient because each run time , we get different importance and coefficient 

This effect means that we cannot precisely draw conclusions about the effects of features on depression level, 

because both feature importance and coefficient tend to fluctuate at each loop time.

```python
import numpy as np
from sklearn.neural_network import MLPClassifier

# Define the number of iterations to perform
n_iter = 1000

# Initialize an empty array to store the feature importances
importance_array = np.zeros((n_iter, len(X.columns)))

# Train and evaluate the MLPClassifier n_iter times
for i in range(n_iter):
    
    X = df.iloc[:,:-1]
    y = df.iloc[:,-1]
    
    # Train an MLPClassifier model
    MLP = MLPClassifier(alpha=0.0001,hidden_layer_sizes=8, max_iter=10000, random_state=i)
    MLP.fit(X, y)
    
    # Get the coefficients of the input layer
    coefficients = MLP.coefs_[0]
    
    feature_names = X.columns
    
    # Calculate the feature importances and coefficients
    importances = np.abs(coefficients).sum(axis=1)
    coefficients_list = np.round(coefficients.ravel(), 3)
    
    # Normalize the importances to sum to 1
    importances /= np.sum(importances)
    
    # Store the importances and coefficients in the array
    importance_array[i, :] = importances
    
# Calculate the mean feature importances and coefficients across all iterations
mean_importances = np.mean(importance_array, axis=0)
mean_coefficients = np.mean(coefficients, axis=1)

# Create a list of (name, importance, coefficient) tuples
feature_importances = list(zip(feature_names, mean_importances, mean_coefficients))

# Sort the list by importance in descending order
feature_importances.sort(key=lambda x: x[1], reverse=True)

# Print the feature importances
for i, (name, importance, coefficient) in enumerate(feature_importances):
    print("{:<5} {:<25} -> avg_importance : {:<10} , avg_coefficient : {}".format(i+1, name, importance.round(3), coefficient.round(3)))
```

From above , as you can see , 

avg_importance of each feature is approximately the same 

-> Each feature affect equally on Depression level

avg_coefficient of each feature is approximately zero 

-> I guess that may be no feature has significant impact on Depression level. (If true, it corresponds to the feature importances ​​computed from the SVC model above.)

So finally I use chi-square test for checking which feature is significant impact on Depression Level or not?

```python
from scipy.stats import chi2_contingency

pair_sig = []
pair_insig = []

for i in range(len(df.columns)-1):
    grouped = X.groupby([df.iloc[:,i], df.iloc[:,-1]]).size().reset_index(name="counts")
    contingency_table = grouped.pivot(index=df.columns[i], columns=df.columns[-1], values="counts").fillna(0).astype(int)

    # Extract the observed frequencies from the contingency table
    observed_frequencies = contingency_table.values

    chi2, p, dof, expected_frequencies = chi2_contingency(observed_frequencies)

    name = df.columns[i] + ' & ' + df.columns[-1]

    if p<0.05:
        print('{:<50} has significant relation with chi-square = {} and p-value = {}'.format(name,chi2,p))
        pair_sig.append(name)
    else:
        print('{:<50} has insignificant relation with chi-square = {} and p-value = {}'.format(name,chi2,p))
        pair_insig.append(name)

    # Create contingency_table

    print('Contingency_table')
    fig, ax = plt.subplots(figsize=(5, 4))
    heatmap = sns.heatmap(contingency_table, annot=True, cmap="GnBu", fmt=".2f", ax=ax)
    
    # Label the rows and columns
    ax.set_xticklabels(contingency_table.columns, rotation=90)
    ax.set_yticklabels(contingency_table.index, rotation=0)
    ax.set_xlabel(df.columns[-1])
    ax.set_ylabel(df.columns[i])
    ax.set_title(name)

    plt.show()

    # Create expected_frequencies
    print('Expected_frequencies')
    fig, ax = plt.subplots(figsize=(5, 4))
    heatmap = sns.heatmap(expected_frequencies, annot=True, cmap="BuPu", fmt=".2f", ax=ax)
    
    # Label the rows and columns
    ax.set_xticklabels(contingency_table.columns, rotation=90)
    ax.set_yticklabels(contingency_table.index, rotation=0)
    ax.set_xlabel(df.columns[-1])
    ax.set_ylabel(df.columns[i])
    ax.set_title(name)

    plt.show()
    print('-'*150)

print('pair_sig : \n {} \n'.format(pair_sig))
print('pair_insig : \n {}'.format(pair_insig))
```

From above , I can conclude that no feature has significant impact on Depression level.
